## Part 2.1: Fourier series 

```python
# Declare the required value to generate the waves.
F = 500.0
Fs = 44100.0
no_samples = 500
T = 1.0 / Fs
```

```python
# Cell to print out graph with harmonics of 5 and Fs / 2
for j in range(1, 3):
    if j == 1:
        harmonics = 100
    else:
        harmonics = int(Fs / 2)

    x_axis =np.zeros(no_samples)
    fourier_a = np.zeros(no_samples)
    fourier_b = np.zeros(no_samples)
    fourier_c = np.zeros(no_samples)

    # For loops to print out 3 fourier transform a,b,c with harmonics = 100.
    for i in range(0, no_samples):
        x_axis[i] = i
        fourier_a[i] = 1.0
        fourier_b[i] = 1.0
        fourier_c[i] = 1.0

        for k in range(0, harmonics):
        # Check for odd only as it's value is not equal 0
            if k % 2 == 1:
                fourier_a[i] = (fourier_a[i] + (np.sin(2 * np.pi * k * F * i * T) * float(1.0 / k)))
                fourier_b[i] = (fourier_b[i] + (np.cos(2 * np.pi * k * F * i * T) * float(1.0 / (k * k))))
                if k != 3:
                    fourier_c[i] = (fourier_c[i] + (np.sin((2 * np.pi * k * F * i * T) + 0) * float(1.0 / k)))
                elif k == 3:
                    fourier_c[i] = (fourier_c[i] + (np.sin((2 * np.pi * k * F * i * T) + (np.pi/2)) * float(1.0 / k)))

    # Plot 3 different Fourier series graph a,b,c
    fig, ax1 = pyplot.subplots(1)
    fig, ax2 = pyplot.subplots(1)
    fig, ax3 = pyplot.subplots(1)
    
    ax1.set_title("Fourier Series A with harmonics = " + str(harmonics))
    ax2.set_title("Fourier Series B with harmonics = " + str(harmonics))
    ax3.set_title("Fourier Series C with harmonics = " + str(harmonics))
    
    ax1.plot(x_axis, fourier_a)
    ax2.plot(x_axis, fourier_b)
    ax3.plot(x_axis, fourier_c)
```

### Part 2.3 Transforming music files to & from frequency-domain

```python
def sin_wave(input_frequency, size, amplitude):
    Fs = 44100.0 # Sampling frequency = 44100.0 (Hz) 
    T = 1.0 / Fs # Sampling period (seconds)
    A = amplitude # Amplitude of sine-wave = 30000
    f = input_frequency # Frequency of sine-wave (Hz)

    # Producing sound wave.
    y = np.zeros(size) ; i = np.arange(0,size)
    y = A * np.sin( 2 * np.pi * T * (i*size) )
    quant_y = np.int16(y) # Convert y to 16-bits
    return quant_y

# Open the required file
x = open("filename.wb", 'wb')
(Fs, y_file) = wavfile.read('SVivaldi44.1mono.wav')
print "Read the file SVivaldi44.1mono.wav"

# Generate a sin wave with f = 100, size = 500, A = 100
y = sin_wave(100.0, 500, 100)
fig, ax = pyplot.subplots(1)
ax.plot(y)
ax.set_title("Generated Sin wave")
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
ax.grid(True)

# Get the magnitude values for different frequencies
y_dct = sp.fftpack.dct(y, norm='ortho')
fig, ax = pyplot.subplots(1)
ax.stem(y_dct)
ax.set_title("Generated dct")
ax.set_xlabel('Frequency')
ax.set_ylabel('Amplitude')
ax.grid(True)

# Get the magnitude values for different frequencies
fig, ax= pyplot.subplots(1)
ax.stem(np.absolute(y_dct))
ax.set_title("Magnitude of dct")
ax.set_xlabel('Frequency Hz')
ax.set_ylabel('Magnitude')
ax.grid(True)

# Undo the dct process and recover the original wave
x = sp.fftpack.idct(y_dct, norm='ortho')
fig, ax = pyplot.subplots(1)
ax.plot(x)
ax.set_title("Recover Sin wave")
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
ax.grid(True)
```

## Part 2.6: Investigating frequency-masking

```python
def dB_SPL(A, A_0):
    return 20 * np.log10(A / A_0)

def spreading_function(freq_mask, freq_test, mask_loud):
    freq_diff = freq_test - freq_mask
    zd = 13 * math.atan(0.00076*freq_diff) + 3.5 * math.atan((freq_diff / 7500.0)**2)
    return mask_loud + 15.81 + 7.5 * (zd + 0.474) - 17.5 * math.sqrt(1+(zd+0.474)**2)    

def sine_wave(A, f, duration) :
    Fs = 44100.0 # Sampling frequency = 44100.0 (Hz) 
    T = 1.0 / Fs # Sampling period (seconds)
    size = int(duration / T) # Total number of x-axix element

    # Producing sound wave.
    y = np.zeros(size) ; i = np.arange(0,size)
    y = A * np.sin( 2 * np.pi * f * i * T )
    
    quant_y = np.int16(y) # Convert y to 16-bits
    Audio(quant_y,Fs) # Output the audio file
    return quant_y
```

```python
# Plot the freq-masking graph using the formula
def spreading_plot(mask_freq, ratio):
    # Use spreading function formula to plot a graph
    A_0 = 20000.0 # Max range is 20 kHz

    x = []

    for n in range(1,20001):
        mask_loud = dB_SPL(A_0, 16)
        x.append(spreading_function(mask_freq, n, mask_loud) * ratio)

    x = np.int16(x)
    return x

fig, ax = pyplot.subplots(1)
ax.set_title("Masking by 1 kHz tone 0.8 time amplitude")
ax.set_xlabel('Frequency Hz')
ax.set_ylabel('dB_SPL')
ax.plot(spreading_plot(1000, 0.8))

# From the graph we can see it's peak at 50
```
