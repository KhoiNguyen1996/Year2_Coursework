## Part 1.1 Generating and listening to sine-waves

```python
# Function to produce an audio file of 3 seconds for a given input_frequency.
def soundWave_44kHz(input_frequency) :
    Fs = 44100.0 # Sampling frequency = 44100.0 (Hz) 
    T = 1.0 / Fs # Sampling period (seconds)
    A = 30000.0 # Amplitude of sine-wave = 30000
    f = input_frequency # Frequency of sine-wave (Hz)
    size = int(3 / T) # Total number of x-axix element

    # Producing sound wave.
    y = np.zeros(size) ; i = np.arange(0,size)
    y = A * np.sin( 2 * np.pi * f * i * T )
    
    quant_y = np.int16(y) # Convert y to 16-bits
    
    return Audio(quant_y,Fs) # Output the audio file
```

```python
soundWave_44kHz(250) # Producing sound with frequency = 250 Hz for 3 seconds.
```

```python
# For loop to produce sounds with frequency from 250-2000 Hz with 250 Hz increment.
for x in range(0,8):
    frequency = x * 250 + 250
    print "For frequency = ", frequency, " Hz"
    soundWave_44kHz(frequency)
```

## Part 1.2. Demonstration of aliasing for sine-waves.

```python
# Function to produce an audio file of 3 seconds
# with Sampling frequency = 4000 Hz and
# amplitude of sine-wave = 30000
# for a given input_frequency.
def soundWave_4kHz(input_frequency) :
    Fs = 4000.0 # Sampling frequency = 4000.0 (Hz) 
    T = 1.0 / Fs # Sampling period (seconds)
    A = 30000.0 # Amplitude of sine-wave = 30000
    f = input_frequency # Frequency of sine-wave (Hz)
    size = int(3 / T) # Total number of x-axis element

    # Producing sound wave.
    y = np.zeros(size) ; i = np.arange(0,size)
    y = A * np.sin( 2 * np.pi * f * i * T )
    
    quant_y = np.int16(y) # Convert y to 16-bits

    return Audio(quant_y,Fs) # Output the audio file
```

```python
# For loop to create sound with frequency from 500 to 5000 Hz for 3 seconds with 500 Hz increment.
for x in range(0,10) :
    frequency = x * 500 + 500
    print "For frequency = ", frequency, " Hz"
    soundWave_4kHz(frequency)
```

## Part 1.5. Processing a music file with different bit-rate and sampling frequency 

```python
# Function to produce an audio file with different bit-rate.
def bitSampling(name, NB) :
    print "File name:", name, "with NB =", NB
    NB = NB - 1
    
    (Fs, y) = wavfile.read(name)
    SM = max(abs(y)) # Get maximum amplitude of the sound wave.
    y = y / float(SM) # Scale the sound wave down to range of -1 to 1.

    # Reducing the number of bit per sample.
    quant_y = np.round(y * (2**NB - 0.5) - 0.5)

    # Increasing the sound volume.
    iquant_y = np.int16((quant_y + 0.5) * 2**(15 - NB))

    return iquant_y
```

```python
# Read music from file.
(Fs, y) = wavfile.read("HQ-music44100-mono.wav")
print "Original music file"
Audio(y, Fs)

# For loop to identify the minimum number of bits per sample for any sampling frequency
# and with the minimum acceptable sampling frequency Fs = 22050 Hz.
for i in range(0,17) :
    print "========================================================="
    y = bitSampling("HQ-music44100-mono.wav", i)
    print "With default sampling frequency"
    Audio(y, Fs)
    reduced_y = int(len(y) / 2)
    yr = resample(y, reduced_y)
    quant_yr = np.int16(yr) # Convert yr to 16-bits
    print "With sampling frequency = 22050 Hz"
    Audio(quant_yr, 22050)
```

## Part 1.7. Log-PCM encoded speech

```python
# Function to compress a given speech's wave array using mu-law. With mu = 255 
def compress(x) :
    mu = 255.0

    # Scaled x to range from -1 to 1.
    SM = max(abs(x)) # Get maximum amplitude of the sound wave.
    x = x / float(SM) # Scale the sound wave down to range of -1 to 1.

    # Uniform quantisation of 16-bits.
    NB = 16 - 1
    quant_x = np.round(x * (2**NB - 0.5) - 0.5)
    
    SM = max(abs(quant_x)) # Get maximum amplitude of the sound wave.
    quant_x = quant_x / float(SM) # Scale the sound wave down to range of -1 to 1.

    # Companding the sound using mu-law
    return (np.sign(quant_x) * (np.log(1.0 + mu * abs(quant_x)) / np.log(1.0 + mu)))

# Function to expand a given speech's wave array using mu-law.
def expand(y, NB) :
    mu = 255.0

    # Uniform quantisation of 8-bits.
    NB = NB - 1
    quant_y = np.round(y * (2**NB - 0.5) - 0.5)
    
    SM = max(abs(quant_y)) # Get maximum amplitude of the sound wave.
    quant_y = quant_y / float(SM) # Scale the sound wave down to range of -1 to 1.

    # Expanding the sound using mu-law
    return np.sign(quant_y) * (((1.0 + mu)**abs(quant_y)) - 1.0) / mu
```

```python
# Plotting expand and compress graph for 200 values from -1 to 1.
# Initialise numpy array from -1 to 1 with 200 values and 0.01 increment.
xValues = np.arange(-1,1.01,0.01)
xValues_Compressed = compress(xValues)

# Plot compressing graph
fig, axs = pyplot.subplots(1) 
axs.stem(xValues, xValues_Compressed)
plt.title('Compressed')
axs.grid(True)
axs.set_xlabel("Input")
axs.set_ylabel("Output")

# Plot uncompressing grapgh
xValues_Decompressed = expand(xValues_Compressed, 8)
fig, axs = pyplot.subplots(1)
axs.stem(xValues_Compressed, xValues_Decompressed)
plt.title('Expand')
axs.grid(True)
axs.set_xlabel("Input")
axs.set_ylabel("Output")
```
